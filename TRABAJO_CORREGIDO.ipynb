# **7. Basic Line/Bar Graph**

```python
# Line Graph the Life Expectancy of Colombia
#colombia.index.get_level_values(1)
fig = px.line(colombia, x = 'year_str', y ='VALUE')

# Change title
fig.update_layout(title = 'Bribery in Colombia',
                  title_font_size = 30,
                  title_x = 0.5)

# Change Axis
fig.update_xaxes(title = 'Dates', title_font_size = 20)
fig.update_yaxes(title = 'Bribery in Colombia', title_font_size = 20)

fig.show()
```

```python
# Bar Graph the Life Expectancy of Colombia
fig = px.bar(colombia, x = "year_str", y='VALUE')

# Change title
fig.update_layout(title = 'Bribery in Colombia',
                  title_font_size = 30,
                  title_x = 0.5,
                  yaxis_range = [0,10])

# Change Axis
fig.update_xaxes(title = 'Dates', title_font_size = 20)
fig.update_yaxes(title = 'Bribery in Colombia', title_font_size = 20)

fig.show()
```

## Two y-axis

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x = colombia['year_str'], y = colombia['Category'], name = 'BRIBERY'),
    secondary_y = False, 
)

fig.add_trace(
    go.Scatter(x = colombia['year_str'], y = colombia['VALUE'], name = 'Pocentage Bribery'),
    secondary_y = True, 
)

fig.show()
```

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Bar(x = colombia['year_str'], y = colombia['Category'], name = 'Bribery'),
    secondary_y = False, 
)

fig.add_trace(
    go.Scatter(x = colombia['year_str'], y = colombia['VALUE'], name = 'Pocentage Bribery'),
    secondary_y = True, 
)

fig.show()
```
