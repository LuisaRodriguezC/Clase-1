# **3. Cleaning**

## Column Names

```python
# Change time for year 
d.rename(columns = {" Burglary":"Bribary"}, inplace=True)
#check 
d.head()
```

## A few more things

```python
#(Optional) Duplicate the dataframe
dd = d
```

```python
# Duplicate year variable
dd["year_str"] = dd["Year"].astype(str)

# Check
dd.head()
```

```python
# Create an index
# ESTO NO ES FUNCIONAL DADO QUE LOS INDICES YA ESTÁN EN LA BASE
#idx = pd.MultiIndex.from_product([dd['country'].unique(),
#                                 range(dd.year.min(),dd.year.max()+1)],
#                                 names = ['country','year'])

# Reindex the dataframe
dd = dd.set_index(['Country','Year'])

# Check
dd.head()
```

# **5. Basic Statistics**

```python
# By variable
dd[['Burglary']].describe()
```

# **6. Filters and Pivots**

## Simple Filters

```python
# Filter by country name and save it
colombia = dd.filter(like='Colombia',axis=0)
print(colombia)
```

```python
# Filter by year and save it
y20 = dd.filter(like='2020',axis=0)
print(y20)
```

## Simple Pivots

```python
# Creating is a dataframe with the time-serie of gdp
ts_gdp = d.pivot(index   = 'Year',
                 columns = 'Country',
                 values  = 'Burglary')
print(ts_gdp)
```

# **7. Basic Line/Bar Graph**

```python
# Line Graph the Life Expectancy of Colombia
#colombia.index.get_level_values(1)
fig = px.line(colombia, x = 'year_str', y ='Bribery')

# Change title
fig.update_layout(title = 'Bribery in Colombia',
                  title_font_size = 30,
                  title_x = 0.5)

# Change Axis
fig.update_xaxes(title = 'Dates', title_font_size = 20)
fig.update_yaxes(title = 'Bribery in Colombia', title_font_size = 20)

fig.show()
```

```python
# Bar Graph the Life Expectancy of Colombia
fig = px.bar(colombia, x = "year_str", y='Bribery')

# Change title
fig.update_layout(title = 'Bribery in Colombia',
                  title_font_size = 30,
                  title_x = 0.5,
                  yaxis_range = [0,10])

# Change Axis
fig.update_xaxes(title = 'Dates', title_font_size = 20)
fig.update_yaxes(title = 'Bribery in Colombia', title_font_size = 20)

fig.show()
```

## Two y-axis

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x = colombia['year_str'], y = colombia['Bribery'], name = 'BRIBERY'),
    secondary_y = False, 
)

fig.add_trace(
    go.Scatter(x = colombia['year_str'], y = colombia['Burglary'], name = 'Pocentage Burglary'),
    secondary_y = True, 
)

fig.show()
```

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Bar(x = colombia['year_str'], y = colombia['Bribery'], name = 'Bribery'),
    secondary_y = False, 
)

fig.add_trace(
    go.Scatter(x = colombia['year_str'], y = colombia['Burglary'], name = 'Pocentage Burglary'),
    secondary_y = True, 
)

fig.show()
```
