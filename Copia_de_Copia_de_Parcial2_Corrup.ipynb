# **7. Basic Line/Bar Graph**

```python
# Graph Burglary in colombia
fig = px.line(colombia, x = 'Year_str', y = 'percentage')
fig.show()
```

```python
#Line Graph the life expectancy of Colombia 
fig = px.line(colombia2, x = "Year_str", y='percentage')

#Change title 
fig.update_layout(title = 'Bulgary in Colombia',
                  title_font_size = 30,
                  title_x = 0.5)

#Change Axis 
fig.update_xaxes(title = 'Years', title_font_size = 20)
fig.update_yaxes(title = 'percentage the Bulgary', title_font_size =20)
```

```python
# Bar Graph the life expectancy of Colombia 
fig = px.bar(colombia2, x = "Year_str", y='percentage')

#Change title 
fig.update_layout(title = 'Bulgary in Colombia',
                  title_font_size = 30,
                  title_x = 0.5,
                  yaxis_range = [50,100])

#Change Axis 
fig.update_xaxes(title = 'Years', title_font_size = 20)
fig.update_yaxes(title = 'percentage the Bulgary', title_font_size =20)
```

## Two y-axis 

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces 
fig.add_trace(
    go.Scatter(x = colombia2['Year_str'], y = colombia2['percentage'], name = 'Bulgary'),
    secondary_y = False,
)
fig.add_trace(
    go.Scatter(x = colombia3['Year_str'], y = colombia3['percentage'], name = 'Corruption: Bribery'),
    secondary_y = True,
)

fig.show
```
